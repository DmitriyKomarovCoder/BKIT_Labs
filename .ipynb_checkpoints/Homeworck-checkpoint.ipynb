# Получение данных от сервиса, генерирующего последовательность чисел

Импортируем библиотеки

```python
import requests
import matplotlib.pyplot as plt
```

```python
url = 'http://127.0.0.1:5000/num/12'
# Вызовем сервис
r = requests.get(url)
r
```

```python
data = r.json()
data, type(data)
```

```python
def make_url(cnt):
    base_url = 'http://127.0.0.1:5000/num/'
    res = base_url + str(cnt)
    return res

def get_data(cnt):
    url = make_url(cnt)
    r = requests.get(url)
    return r.json()
```

```python
cnt_list = [5, 10, 15, 100]
for cnt in cnt_list:
    print('{} первых чисел последовательности Фибоначчи: {}'.format(cnt, get_data(cnt)))
```

# Построение графика

```python
y_10 = get_data(10)
x_10 = list(range(1, len(y_10)+1))
```

```python
fig = plt.figure(figsize = (10, 5))
plt.bar(x_10, y_10)
plt.xlabel('Ось абсцисс')
plt.ylabel('Ось ординат')
plt.title('Первые {} чисел последовательности Фибоначчи'.format(len(y_10)))
plt.show()
```
